# Compare features to reference variable

```python
import pandas as pd
from feature_engine.creation import RelativeFeatures
from sklearn.datasets import load_breast_cancer
```

```python
# load the breast cancer dataset from sklearn
data = load_breast_cancer()

# create a dataframe with the independent variables
df = pd.DataFrame(data.data, columns=data.feature_names)

df.head()
```

```python
# # print description of dataset

# print(data.DESCR)
```

```python
# Difference between 2 features - method 1

df["difference"] = df["worst compactness"].sub(df["mean compactness"])

df["difference"].head()
```

```python
# Difference between 2 features - method 2

df["difference"] = df["worst compactness"] - (df["mean compactness"])

df["difference"].head()
```

```python
# Quotient of features - method 1

df["quotient"] = df["worst radius"].div(df["mean radius"])

df["quotient"].head()
```

```python
# Quotient of features - method 2

df["quotient"] = df["worst radius"] / (df["mean radius"])
df["quotient"].head()
```

```python
# features of interest
features = ["mean smoothness", "mean compactness", "mean concavity", "mean symmetry"]

# reference features
reference = ["mean radius", "mean area"]
```

```python
# combine multiple variables with multiple references

creator = RelativeFeatures(
    variables=features,
    reference=reference,
    func=["sub", "div"],
)

df_t = creator.fit_transform(df)
```

```python
# Name of new features

new_features = [f for f in df_t.columns if f not in creator.feature_names_in_]

new_features
```
