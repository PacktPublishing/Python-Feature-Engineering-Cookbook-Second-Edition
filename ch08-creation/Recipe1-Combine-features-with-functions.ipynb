# Combine features with functions

```python
import pandas as pd
from feature_engine.creation import MathFeatures
from sklearn.datasets import load_breast_cancer
```

```python
# load the breast cancer dataset from sklearn
data = load_breast_cancer()

# create a dataframe with the independent variables
df = pd.DataFrame(data.data, columns=data.feature_names)

df.head()
```

```python
# # print description of dataset

# print(data.DESCR)
```

```python
# make a list with the features we want to combine

features = [
    "mean smoothness",
    "mean compactness",
    "mean concavity",
    "mean concave points",
    "mean symmetry",
]
```

```python
df[features].head()
```

```python
# take the average

df["mean_features"] = df[features].mean(axis=1)

df["mean_features"].head()
```

```python
# calculate the standard deviation

df["std_features"] = df[features].std(axis=1)

df["std_features"].head()
```

```python
# list with functions to apply for combinations

math_func = ["sum", "prod", "mean", "std", "max", "min"]
```

```python
# Perform all the operations in one line

df_t = df[features].agg(math_func, axis="columns")

df_t.head()
```

```python
# name of new features

new_feature_names = ["sum_f", "prod_f", "mean_f", "std_f", "max_f", "min_f"]
```

```python
# automate feature combination with Feature-engine

create = MathFeatures(
    variables=features,
    func=math_func,
    new_variables_names=new_feature_names,
)
```

```python
# combine features

df_t = create.fit_transform(df)
```
