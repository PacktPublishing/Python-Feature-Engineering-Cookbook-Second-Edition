```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 2 * np.pi, 0.1)  # start,stop,step
y = np.sin(x)
plt.plot(x, y)

locs = [0, 3, 6]
new_xticks = ["0", "Pi", "2Pi"]
plt.xticks(locs, new_xticks)
plt.ylabel("y")
plt.title("y = sin(x)")
plt.show()
```

```python
x = np.arange(0, 2 * np.pi, 0.1)  # start,stop,step
y = np.cos(x)
plt.plot(x, y)

locs = [0, 3, 6]
new_xticks = ["0", "Pi", "2Pi"]
plt.xticks(locs, new_xticks)
plt.ylabel("y")
plt.title("y = cos(x)")
plt.show()
```
